Drawing A Quantum Graph of An Atom Of Helium.

Helium is a two-electron system, and its quantum states are more complex compared to hydrogen.

![Vast new reserves of helium discovered.jpeg](https://github.com/Lemniscate-SHA-256/Learning-Very-Interesting-Physics-Only/blob/main/Vast%20new%20reserves%20of%20helium%20discovered.jpeg)

Let define the quantum states :

- Electronic States
    - Ground State : $1s^2$
    - Excited States involve one or both electrons in higher orbitals : $1s2s, 1s2p, ...$

I think that learning visually and interactively is more powerful and engaging than articulating flat theorems scribbled on paper like indecipherable hieroglyphics, so we will build a program to visualise atoms based on user input: ground state, excited states and quantum states.
