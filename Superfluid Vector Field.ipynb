Drawing A Superfluid Vector Field

Steps

- Defining vector field equations
- Use Numpy to create a grid for points in space
- Calculate the vector field at each of the grid
- Plot the vector field using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Dfinir la grille de points
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# Dfinir les composants du champ de vecteurs
# Exemple simple: champ de vecteurs radial autour de l'origine
U = -Y / (X**2 + Y**2)
V = X / (X**2 + Y**2)

# Grer les valeurs infinies ou NaN au centre
U[np.isnan(U)] = 0
V[np.isnan(V)] = 0
U[np.isinf(U)] = 0
V[np.isinf(V)] = 0

# Tracer le champ de vecteurs
plt.figure(figsize=(10, 10))
plt.quiver(X, Y, U, V, color='b', pivot='middle', scale=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Champ de vecteurs d\'un superfluide')
plt.grid()

# Tracer les lignes de courant
start_points = np.array([[-5, 5], [5, 5], [-5, -5], [5, -5], [0, 0]])
for start_point in start_points:
    plt.streamplot(x, y, U, V, start_points=start_point.reshape(1, -1), color='r', linewidth=1.5)

plt.show()
```
